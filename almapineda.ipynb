```python
# Specify the username here!
username = 'ornicar'
```

```python
# Importing libraries needed here!
import requests # Handles requests from a website
from bs4 import BeautifulSoup # Web scraping tools
import re # Regex library
```

```python
# Create custom url for specified user
url = 'https://github.com/' + username
# Get the page source code for the url
response = requests.get(url)
```

```python
# check that the connection is viable -- 200 means you're good to go!
response
```

```python
# Create the html parser
soup = BeautifulSoup(response.text, "html.parser")
```

```python
# Find all the info contained within the <rect> tags!
data = soup.findAll('rect')
```

```python
# View data!
data
```

```python
# Function takes in raw data and cleans out unecessary stuff
# and leaves only the contribution amount, date, 
def clean_data(rawdata):
    data_count = []
    count = 0
    for i in data:
        # data_count.append(re.split('"', str(i)))
        c = re.split('"', str(i)) #3, 5, 7
        c_array = [c[3], c[5], c[7]]
        data_count.append(c_array)
    return data_count

data_count = clean_data(data)
```

```python
def longest_streak(data):
    longest = 0;
    count = 0;
    for i in data:
        if int(i[0]) == 0:
            longest = max(count, longest);
            count = 0;
        else:
            count+=1;
    return longest

longest_streak(data_count)
```

```python
def current_streak(data):
    streak = 0;
    for i in range(len(data)-2, -1, -1):
        if int(data[i][0]) == 0:
            return streak;
        else:
            streak+=1;
    return streak

current_streak(data_count)
```
